```python
#import libraires
import pandas as pd
import requests 
from bs4 import BeautifulSoup


url = 'https://www.guessfactory.com/us/en/sale/women/view-all'
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}

result = requests.get(url,headers = headers)
doc = BeautifulSoup(result.text,'html.parser')
prod = doc.find_all('div', {'class': 'product-grid__column})

df = pd.DataFrame(columns = ['Product Number' ,'Product Name', 'New Price', 'Old Price', 'Discount'])
for i in range(40):
    #href = prod[i].find('a')['href'].split('/')
    #no = href[len(href)-1]
    no = no+1
    name = prod[i].find('a', {'class': 'link product-tile__link'}).text.strip() 
    nPrice = prod[i].find('span', {'class': 'price__red-color text-nowrap'}).text.strip()   #new price
    oPrice = prod[i].find('span', {'class': 'price__strike-through'}).text.strip()  #Old price
    discount = prod[i].find('div', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip()  #Old price
    
    new_row = {'Product Number':no, 
               'Product Name':name, 
               'New Price':nPrice,
               'Old Price':oPrice, 
               'Discount':discount}

    df = df.append(new_row,ignore_index=True)
```
