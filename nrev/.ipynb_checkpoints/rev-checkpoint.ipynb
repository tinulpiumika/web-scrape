```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://odel.lk/deal-products'
```

```python
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
```

```python
result = requests.get(url, headers = headers)
```

```python
result
```

```python
doc = BeautifulSoup(result.text,'html.parser')
```

```python
products = doc.findAll('div',{'class','block2'})
```

```python
products[0]
```

```python
href  = products[0].find('a')['href'].split('/')
#product number
print(href[len(href)-1]) 

#product name
print(products[0].find('a',{'class','stext-104'}).text.strip())

#old price
print(products[0].find('del').text.strip())

#new price
print(products[0].find('span',{'class','stext-105'}).text.strip())

#dicount
print(products[0].find('div',{'class','product_tag_discount'}).text.strip())
```

```python
df = pd.DataFrame(columns=['No', 'Product Number', 'Product Name', 'Old Value', 'New Value', 'Discount'])
df
```

```python
i=0
for prod in products:
    i=i+1
    href  = prod.find('a')['href'].split('/')
    prodNo = href[len(href)-1]
    prodName = prod.find('a',{'class','stext-104'}).text.strip()
    oldValue = prod.find('del').text.strip()
    newValue = prod.find('span',{'class','stext-105'}).text.strip()
    discount = prod.find('div',{'class','product_tag_discount'}).text.strip()
    
    row_df = pd.DataFrame([[i, prodNo, prodName, oldValue, newValue, discount]], columns=['No', 'Product Number', 'Product Name', 'Old Value', 'New Value', 'Discount'])
    df = pd.concat([df, row_df],ignore_index=True)
```

```python
df
```

```python
df.to_excel('odelrevoutput.xlsx', index=False)
```
