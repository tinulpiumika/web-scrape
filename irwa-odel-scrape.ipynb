```python
#import libraires
import pandas as pd
import requests 
from bs4 import BeautifulSoup

url = 'https://odel.lk/deal-products'
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}

result = requests.get(url,headers = headers)
doc = BeautifulSoup(result.text,'html.parser')
prod = doc.findAll(class_='block2')

df = pd.DataFrame(columns = ['Product Number' ,'Product Name', 'New Price', 'Old Price', 'Discount'])
for i in range(40):
    href = prod[i].find('a')['href'].split('/')
    no = href[len(href)-1]
    name = prod[i].find(class_='stext-104').text.strip()
    nPrice = prod[i].find(class_='stext-105').text.strip()  #new price
    oPrice = prod[i].find('del').text.strip() #Old price
    discount = prod[i].find(class_='product_tag_discount').text.strip()  #Old price
    
    new_row = {'Product Number':no, 
               'Product Name':name, 
               'New Price':nPrice,
               'Old Price':oPrice, 
               'Discount':discount}

    df = df.append(new_row,ignore_index=True)
```

```python
#import libraires
import pandas as pd
import requests 
from bs4 import BeautifulSoup

url = 'https://odel.lk/deal-products'
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
}
```

```python
result = requests.get(url,headers = headers)
doc = BeautifulSoup(result.text,'html.parser')
products = doc.findAll(class_='block2')
df = pd.DataFrame(columns = ['No','Product Number' ,'Product Name', 'New Price', 'Old Price', 'Discount'])
i=1
for prods in products:
    href = prods.find('a')['href'].split('/')
    no = href[len(href)-1]
    name = prods.find(class_='stext-104').text.strip()
    nPrice = prods.find(class_='stext-105').text.strip()  #new price
    oPrice = prods.find('del').text.strip() #Old price
    discount = prods.find(class_='product_tag_discount').text.strip()  #Old price
    
    ndf = pd.DataFrame([[i,no, name, nPrice, oPrice, discount]], columns = ['No','Product Number' ,'Product Name', 'New Price', 'Old Price', 'Discount'])

    df = pd.concat([df,ndf],ignore_index=True)
    i=i+1
```
